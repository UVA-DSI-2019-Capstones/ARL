```python
master_list = []
def h(j, df1, df2):
    value_list = []
    df_new = pd.DataFrame(columns = df1.columns)
    for i in range(0,df1.shape[0]):
        for a in range(0,df2.shape[0]):
            if not isinstance(df2.iloc[a,j], float) and not isinstance(df1.iloc[i,j], float):
                value_list.append(df1.iloc[i,j] + df2.iloc[a,j]) 
    master_list.append(value_list)
    print(j)
    if j+1 != df1.shape[1]:
        h(j+1, df1, df2)
    return master_list
```
