```python
from docx import Document
def getText(filename):
    doc = Document(filename)
    fullText = []
    for para in doc.paragraphs:
          fullText.append(para.text)
    return fullText
```

```python
file = open('correct_listwav.txt', 'r')

pre_file = []
for line in file:
    line = line.split('\n')[0]
    if 'pre' in line.casefold():
        pre_file.append(line)
del(pre_file[6])
#Put this back in later
del(pre_file[pre_file.index('9513-PRE.wav')])
del(pre_file[pre_file.index('9549-PREtest.wav')])
```

```python
column_list = list(pre_data.columns)
```

```python
for k in range(0,len(pre_file)):
    text_data = getText('labeled_data\\' + pre_file[k]+'.docx')
    print(k)
    t_text_data = []
    for p in range(0,len(text_data)):
        try:
            if text_data[p][0] == 'T':
                if '~' in text_data[p][2:-1]:
                    t_text_data.append(text_data[p].split(':')[1].split('~'))
        except:
            continue

    for i in range(0,len(t_text_data)):
        label = t_text_data[i][0].split(',')
        for j in range(0,len(label)):
            row = int(label[j])*2+1
            if pre_data.iloc[row, column_list.index(pre_file[k][0:4])] == '0':
                pre_data.iloc[row, column_list.index(pre_file[k][0:4])] = t_text_data[i][1]
            else:
                pre_data.iloc[row, column_list.index(pre_file[k][0:4])] = pre_data.iloc[row, column_list.index(pre_file[k][0:4])] + ' ~ ' + t_text_data[i][1]
```

```python
pre_data = pre_data.iloc[:,:-2]
```
